```python
group_data = df.groupby(
    ['title', 'year', 'version', 'type']
).agg(
    {
        'director': 'last',
        'actor': 'last',
        'age': 'first',
        'runtime': 'first',
        'rating': 'first',
        'like': 'last',
        'comment': 'last',
        'isDRM': 'first',
        'cost': 'last',
        'period': 'first',
        'is_action': 'sum',
        'is_comedy': 'sum',
        'is_drama': 'sum',
        'is_melo': 'sum',
        'is_horror': 'sum',
        'is_fantasy': 'sum',
        'is_animation': 'sum',
        'is_DQ': 'sum',
        'is_indie': 'sum',
        'is_concert': 'sum',
    }
).reset_index()
```

```python
group_data.info()
```

```python
grouped_data = group_data.groupby(['title', 'year', 'actor', 'director']).size()

frequent_titles = grouped_data[grouped_data >= 3].reset_index(name='count')['title'].tolist()
len(frequent_titles)
```
