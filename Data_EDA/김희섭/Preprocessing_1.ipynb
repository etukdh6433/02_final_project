```python
df = pd.read_csv('시리즈온_영화_데이터.csv')
```

```python
df.fillna('Null', inplace=True)

group_data = df.groupby(
    ['title', 'year', 'director', 'actor', 'version', 'type']
).agg(
    {
        'age': 'first',
        'runtime': 'first',
        'rating': 'first',
        'like': 'last',
        'comment': 'last',
        'isDRM': 'first',
        'cost': 'last',
        'period': 'first',
        'is_action': 'sum',
        'is_comedy': 'sum',
        'is_drama': 'sum',
        'is_melo': 'sum',
        'is_horror': 'sum',
        'is_fantasy': 'sum',
        'is_animation': 'sum',
        'is_DQ': 'sum',
        'is_indie': 'sum',
        'is_concert': 'sum',
    }
).reset_index()
```
